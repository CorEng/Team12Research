```python
## Import 145 Pairs
pairs145 = [line.rstrip('\n') for line in open("145pairs.txt")]
```

```python
import re
commalist = []
for i in pairs145:
    commalist.append(re.sub("[ ]", ",", i))
```

```python
dividedlist = []
for i in commalist:
    dividedlist.append(i.split(","))
    
```

```python
uniquelist = []
for i in dividedlist:
    if (i[0] +","+ i[1]) or (i[1] +","+ i[0]) not in uniquelist:
        uniquelist.append(i[0] +","+ i[1])
```

```python
stopdic = {}
for index, row in stopdata.iterrows():
    stopdic[str(row["stopid"])] = (str(row["lat"]) + "," + str(row["lng"])).split(",")
```

```python
uniquecoordpairs = {}
for i in dividedlist:
    try:
        uniquecoordpairs[str(i)]=str(stopdic[str(i[0])]) + ":" + str(stopdic[str(i[1])])
    except:
        uniquecoordpairs[str(i)]="FLAG"
```

```python
## Start getting distance value for pairs

import pandas as pd
import googlemaps
from itertools import tee
```

```python
API_key = ''#enter Google Maps API key
gmaps = googlemaps.Client(key=API_key)
```

```python
for i in uniquecoordpairs:
    print(str(uniquecoordpairs[str(i)].split(":")[1]).replace("'", '').replace("[","").replace("]",""))
```

```python
distancedic ={}
for i in uniquecoordpairs:
    try:
        begin = str(uniquecoordpairs[str(i)].split(":")[1]).replace("'", '').replace("[","").replace("]","")
        end = str(uniquecoordpairs[str(i)].split(":")[0]).replace("'", '').replace("[","").replace("]","")
        distancedic[str(i)] = gmaps.distance_matrix(begin, end, mode='walking')["rows"][0]["elements"][0]["distance"]["value"]
    except:
        distancedic[str(i)] = "FLAG"
```
